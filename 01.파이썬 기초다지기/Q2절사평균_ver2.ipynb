```python
# 깃허브의 www.github.com/hyokwan -> python_lecture 클릭 dataset ->
# kopo_customerdata.csv
csFilePath = "https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/kopo_customerdata.csv"
customerData = pd.read_csv(csFilePath)
# customerTotalList = list(csData.TOTAL_AMOUNT)
customerTotalList = customerData["TOTAL_AMOUNT"].tolist()
customerTotalList
```

```python
# Step1. 리스트 선언
#customerTotalList

# Step2. 최소값, 최대값 계산
minValue = min(customerTotalList)
maxValue = max(customerTotalList)

# Step3. 최소값, 최대값 개수
cntMin = customerTotalList.count(minValue)
cntMax = customerTotalList.count(maxValue)

# Step4. 리스트 오름차순 정렬하기
customerTotalList.sort()
```

```python
#Step5. 최소값, 최대값 "모두" 제거한 리스트
rmMinMaxList = customerTotalList[cntMin:-cntMax]
```

```python
#Step6. 평균 계산(전체합/개수)
sumExceptList = sum(rmMinMaxList)
cntExceptList = len(rmMinMaxList)
avgExceptList = sumExceptList / cntExceptList
avgExceptList
```
